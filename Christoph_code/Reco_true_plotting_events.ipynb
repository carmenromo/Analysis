```python
import os
import pandas as pd
import numpy  as np

from invisible_cities.io.dst_io import load_dst
import matplotlib.pyplot as plt
```

```python
hdf_dir  = '/Users/carmenromoluque/nexus_petit_analysis/PETit-ring/Christoff_sim/compton/def_files_Chr/0.0.0/'
#filename = hdf_dir + 'full_ring_irad165_p7mm2_info_0.0.0cm_000.h5'
#filename = hdf_dir + 'full_ring_irad165_p7mm2_info_coinc_0.0.0cm_000.h5'
filename = hdf_dir + 'full_ring_irad165_p7mm2_info_singles_0.0.0cm_001.h5'
dst = load_dst(filename, "DST", "Events")
```

## True info

```python
true_x1 = dst.True_x1.values
true_y1 = dst.True_y1.values
true_x2 = dst.True_x2.values
true_y2 = dst.True_y2.values
true_x  = np.concatenate((true_x1, true_x2), axis=0)
true_y  = np.concatenate((true_y1, true_y2), axis=0)
plt.figure(figsize=(10,10))
plt.plot(true_x, true_y, '.', markersize=1)
plt.xlabel('True x', size=18)
plt.ylabel('True y', size=18)
```

## Reco info

```python
reco_x1 = dst.Reco_x1.values
reco_y1 = dst.Reco_y1.values
reco_x2 = dst.Reco_x2.values
reco_y2 = dst.Reco_y2.values
reco_x  = np.concatenate((reco_x1, reco_x2), axis=0)
reco_y  = np.concatenate((reco_y1, reco_y2), axis=0)
plt.figure(figsize=(10,10))
plt.plot(reco_x1, reco_y1, '.', markersize=1)
plt.xlabel('Reco x1', size=18)
plt.ylabel('Reco y1', size=18)
```

## Diff reco-true

```python
out = 0
plt.figure(figsize=(10,10))
for (tx1, ty1, rx1, ry1) in zip(true_x1[:3000], true_y1[:3000], reco_x1[:3000], reco_y1[:3000]):
    xs = np.array([tx1, rx1])
    ys = np.array([ty1, ry1])
    dist = np.sqrt((tx1-rx1)**2 + (ty1-ry1)**2)
    if dist > 50:
        out += 1
    plt.plot(xs, ys, '.', markersize=0.5, linestyle='-', linewidth=0.5, c='r')
print(out)
```
