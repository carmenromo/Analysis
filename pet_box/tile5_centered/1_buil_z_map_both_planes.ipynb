### Store data for the table:

```python
varXnBins   = [50, 100, 150, 200]
varXRange   = (0, 210)
varXBins    = {n: np.linspace(*varXRange, n + 1) for n       in varXnBins}
varXCenters = {n: shift_to_bin_centers(bins)     for n, bins in varXBins.items()}
varXPitch   = {n: np.diff(bins)[0]               for n, bins in varXBins.items()}
```

```python
thr  = 2
sel2      = in_range(true_z1_a[thr][sel], *z_lim) & in_range(var_x1_a[thr][sel], *varx_lim)
sel_lines = (var_x1_a [thr][sel] < (line_test)) & (var_x1_a [thr][sel] > (line_test2))
z, v      = true_z1_a[thr][sel][sel2&sel_lines], var_x1_a[thr][sel][sel2&sel_lines]

Z_ave   = {}
Z_ave_u = {} # uncertainty
for n in varXnBins:
    x, y, yu   = fitf.profileX(v, z, n, varXRange, drop_nan=False)
    Z_ave  [n] = y
    Z_ave_u[n] = yu
    print(x, y, yu)
```

### Create table

```python
path = '/Users/carmenromoluque/nexus_petit_analysis/tof_setup/PetBox_analysis/'
varx_table_name = path+'z_var_x_table_pet_box_HamamatsuVUV_det_plane_coinc_plane_cent.h5'

class ZposT(tb.IsDescription):
    Var_x           = tb.Float32Col(pos=0)
    Zpos            = tb.Float32Col(pos=0)
    ZposUncertainty = tb.Float32Col(pos=0)

with tb.open_file(varx_table_name, "w") as output_file:
    
    if '/Zpos' in output_file:
        Zpos = output_file.root.Zpos        
    else:
        Zpos = output_file.create_group(output_file.root, "Zpos")
    
    for n in varXBins:
        pitch      = varXPitch[n]
        table_name = f"f{thr}pes{n}bins"

        z_table = output_file.create_table(Zpos, table_name,
                               description = ZposT,
                               title       = "z_pos",
                               filters     = tbl.filters('ZLIB4'))
    
    
        for i, var in enumerate(varXCenters[n], start=0):
            row                    = z_table.row
            row["Var_x"]           = var
            row["Zpos"]            = Z_ave  [n][i]
            row["ZposUncertainty"] = Z_ave_u[n][i] * Z_ave[n][i] / 100
            row.append()

        table = getattr(getattr(output_file.root, 'Zpos'), table_name)
        table.attrs.varX_bins  = n
        table.attrs.varX_pitch = pitch
```

```python
path = '/Users/carmenromoluque/nexus_petit_analysis/tof_setup/PetBox_analysis/'
varx_table_name = path+'z_var_x_table_pet_box_HamamatsuVUV_coinc_plane_cent.h5'

class ZposT(tb.IsDescription):
    Var_x           = tb.Float32Col(pos=0)
    Zpos            = tb.Float32Col(pos=0)
    ZposUncertainty = tb.Float32Col(pos=0)

with tb.open_file(varx_table_name, "w") as output_file:
    
    if '/Zpos' in output_file:
        Zpos = output_file.root.Zpos        
    else:
        Zpos = output_file.create_group(output_file.root, "Zpos")
    
    for n in varXBins:
        pitch      = varXPitch[n]
        table_name = f"f{thr}pes{n}bins"

        z_table = output_file.create_table(Zpos, table_name,
                               description = ZposT,
                               title       = "z_pos",
                               filters     = tbl.filters('ZLIB4'))
    
    
        for i, var in enumerate(varXCenters[n], start=0):
            row                    = z_table.row
            row["Var_x"]           = var
            row["Zpos"]            = Z_ave  [n][i]
            row["ZposUncertainty"] = Z_ave_u[n][i] * Z_ave[n][i] / 100
            row.append()

        table = getattr(getattr(output_file.root, 'Zpos'), table_name)
        table.attrs.varX_bins  = n
        table.attrs.varX_pitch = pitch
```
