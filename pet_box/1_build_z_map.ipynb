```python
import os
import numpy             as np
import pandas            as pd
import tables            as tb
import matplotlib.pyplot as plt

%matplotlib inline

#import invisible_cities.core .fit_functions  as     fitf
#from   invisible_cities.core .core_functions import in_range
from   invisible_cities.core.core_functions  import shift_to_bin_centers
#from   invisible_cities.reco                 import tbl_functions as tbl
```

```python
def create_new_figure(kwargs):
    if kwargs.setdefault("new_figure", True):
        plt.figure()
    del kwargs["new_figure"]
    
def hist2d(*args, **kwargs):
    """
    Create a figure and then the histogram
    """
    create_new_figure(kwargs)

    z, x, y, p = plt.hist2d(*args, **kwargs)
    return z, shift_to_bin_centers(x), shift_to_bin_centers(y), p
```

```python
eventsPath = '/Users/carmenromoluque/nexus_petit_analysis/tof_setup/PetBox_analysis/data_z_map/'

thr_ch_start  = 0
thr_ch_nsteps = 5

true_z = [[] for i in range(thr_ch_start, thr_ch_nsteps)]
var_r  = [[] for i in range(thr_ch_start, thr_ch_nsteps)]

for filename in os.listdir(eventsPath):
    if filename.endswith('10_0_5.npz'):
        my_file   = eventsPath+filename
        d = np.load(my_file, allow_pickle=True)
        for j in range(thr_ch_start, thr_ch_nsteps):
            true_z[j] = np.concatenate((np.array(d['a_true_z'][j]), true_z[j]))
            var_r [j] = np.concatenate((np.array(d['a_var_r'] [j]), var_r [j]))

true_z = np.array(true_z)
var_r  = np.array(var_r )
```

```python
for i, th in enumerate(range(thr_ch_start, thr_ch_nsteps)):
    print(len(true_z[i]), len(var_r[i]))
    print(max(true_z[i]), max(var_r[i]))
    print(min(true_z[i]), min(var_r[i]))
```

```python
ranges = (27., 54.), (0., 120)
bins   = 200

for i, th in enumerate(range(thr_ch_start, thr_ch_nsteps)):
    plt.rcParams["font.size"]      = 12
    plt.rcParams["figure.figsize"] = 8, 6

    h1 = hist2d(-true_z[i], var_r[i], bins=(bins, bins), range=ranges, cmin=1)

    plt.xlabel('Z pos (mm)')
    plt.ylabel('Var r (mm)')
    plt.title(f'Thr = {i}')
    plt.colorbar().set_label("Number of events")
    
    plt.tight_layout()
    plt.show()
```
