## Plot contrast recovery coefficient (CRC) as a function of the activity

### Image parameters

```python
sig_intensity  =  4
bckg_intensity =  1
radial_pos     = 50.
bckg_radius    =  4.
phi0           = np.pi/6
phi_step       = np.pi/3
n_phi          = 6

hot_radii   = [4., 6.5, 8.5, 11.]
cold_radii  = [14., 18.5]
hot_angles  = [np.pi/3, 2*np.pi/3, 3*np.pi/3, 4*np.pi/3]
cold_angles = [5*np.pi/3, 6*np.pi/3]

x_size = y_size = z_size = 180
x_bins = y_bins = z_bins =  60
```

```python
all_imgs = [the_imgs0, the_imgs1, the_imgs2, the_imgs3, the_imgs4]
```

```python
it = 7 # would mean 16

crcs_hot_0, crcs_hot_1   = [], []
crcs_hot_2, crcs_hot_3   = [], []
crcs_cold_0, crcs_cold_1 = [], []

err_crcs_hot_0, err_crcs_hot_1   = [], []
err_crcs_hot_2, err_crcs_hot_3   = [], []
err_crcs_cold_0, err_crcs_cold_1 = [], []

for i in range(len(all_imgs)):
    crc_hot0, err_hot0 = fomf.crc_hot3d(all_imgs[i][it], sig_intensity, bckg_intensity, hot_radii[0], radial_pos,
                                        hot_angles[0], bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                        x_bins, y_bins, z_bins)
    crc_hot1, err_hot1 = fomf.crc_hot3d(all_imgs[i][it], sig_intensity, bckg_intensity, hot_radii[1], radial_pos,
                                        hot_angles[1], bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                        x_bins, y_bins, z_bins)
    crc_hot2, err_hot2 = fomf.crc_hot3d(all_imgs[i][it], sig_intensity, bckg_intensity, hot_radii[2], radial_pos,
                                        hot_angles[2], bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                        x_bins, y_bins, z_bins)
    crc_hot3, err_hot3 = fomf.crc_hot3d(all_imgs[i][it], sig_intensity, bckg_intensity, hot_radii[3], radial_pos,
                                        hot_angles[3], bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                        x_bins, y_bins, z_bins)
    crcs_hot_0.append(crc_hot0*100)
    crcs_hot_1.append(crc_hot1*100)
    crcs_hot_2.append(crc_hot2*100)
    crcs_hot_3.append(crc_hot3*100)
    
    err_crcs_hot_0.append(err_hot0*100)
    err_crcs_hot_1.append(err_hot1*100)
    err_crcs_hot_2.append(err_hot2*100)
    err_crcs_hot_3.append(err_hot3*100)
    
    crc_cold0, err_cold0 = fomf.crc_cold3d(all_imgs[i][it], cold_radii[0], radial_pos, cold_angles[0], 
                                           bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                           x_bins, y_bins, z_bins)
    crc_cold1, err_cold1 = fomf.crc_cold3d(all_imgs[i][it], cold_radii[1], radial_pos, cold_angles[1], 
                                           bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                           x_bins, y_bins, z_bins)
    crcs_cold_0.append(crc_cold0*100)
    crcs_cold_1.append(crc_cold1*100)
    
    err_crcs_cold_0.append(err_cold0*100)
    err_crcs_cold_1.append(err_cold1*100)
```

```python
nc = np.array(n_coincidences)/1e6
ls = [f'R =  {hot_radii[0]} (hot sphere)', f'R =  {hot_radii[1]} (hot sphere)', f'R =  {hot_radii[2]} (hot sphere)',
      f'R = {hot_radii[3]} (hot sphere)', f'R = {cold_radii[0]} (cold sphere)', f'R = {cold_radii[1]} (cold sphere)']
plt.errorbar(nc, crcs_hot_0,  yerr=err_crcs_hot_0,  xerr=None, label=ls[0], color=light_red,  marker='s', capsize=3)
plt.errorbar(nc, crcs_hot_1,  yerr=err_crcs_hot_1,  xerr=None, label=ls[1], color=light_blue, marker='s', capsize=3)
plt.errorbar(nc, crcs_hot_2,  yerr=err_crcs_hot_2,  xerr=None, label=ls[2], color=blue,     marker='s', capsize=3)
plt.errorbar(nc, crcs_hot_3,  yerr=err_crcs_hot_3,  xerr=None, label=ls[3], color=green,    marker='s', capsize=3)
plt.errorbar(nc, crcs_cold_0, yerr=err_crcs_cold_0, xerr=None, label=ls[4], color='k',      marker='s', capsize=3)
plt.errorbar(nc, crcs_cold_1, yerr=err_crcs_cold_1, xerr=None, label=ls[5], color='orange', marker='s', capsize=3)
plt.title('CRC vs # coincidences (thr=0.5 pes)')
plt.xlabel(f'Number of coincidences used (x $10^6$)')
plt.ylabel('CRC (%)')
plt.ylim(0, 90)
plt.legend(loc=9, ncol=2);
```
