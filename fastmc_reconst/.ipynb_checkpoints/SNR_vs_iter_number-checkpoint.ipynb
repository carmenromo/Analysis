```python
n_iterations = 100
save_every   = 2

the_imgs = []
for i in range(0, n_iterations, save_every):
    img = rec.read_image(i)
    the_imgs.append(img)
```

### Image parameters

```python
sig_intensity  =  4
bckg_intensity =  1
radial_pos     = 50.
bckg_radius    =  4.
phi0           = np.pi/6
phi_step       = np.pi/3
n_phi          = 6

hot_radii   = [4., 6.5, 8.5, 11.]
cold_radii  = [14., 18.5]
hot_angles  = [np.pi/3, 2*np.pi/3, 3*np.pi/3, 4*np.pi/3]
cold_angles = [5*np.pi/3, 6*np.pi/3]

x_size = y_size = z_size = 180
x_bins = y_bins = z_bins =  60
```

```python
snrs_hot_0,  snrs_hot_1  = [], []
snrs_hot_2,  snrs_hot_3  = [], []
snrs_cold_0, snrs_cold_1 = [], []

err_snrs_hot_0,  err_snrs_hot_1  = [], []
err_snrs_hot_2,  err_snrs_hot_3  = [], []
err_snrs_cold_0, err_snrs_cold_1 = [], []

for i in range(len(the_imgs)):
    snr_hot0, err_hot0 = fomf.snr3d(the_imgs[i], hot_radii[0], radial_pos, hot_angles[0], 
                                bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                x_bins, y_bins, z_bins)
    snr_hot1, err_hot1 = fomf.snr3d(the_imgs[i], hot_radii[1], radial_pos, hot_angles[1], 
                                bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                x_bins, y_bins, z_bins)
    snr_hot2, err_hot2 = fomf.snr3d(the_imgs[i], hot_radii[2], radial_pos, hot_angles[2], 
                                bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                x_bins, y_bins, z_bins)
    snr_hot3, err_hot3 = fomf.snr3d(the_imgs[i], hot_radii[3], radial_pos, hot_angles[3], 
                                bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                x_bins, y_bins, z_bins)

    snrs_hot_0.append(snr_hot0*100)
    snrs_hot_1.append(snr_hot1*100)
    snrs_hot_2.append(snr_hot2*100)
    snrs_hot_3.append(snr_hot3*100)
    
    err_snrs_hot_0.append(err_hot0*100)
    err_snrs_hot_1.append(err_hot1*100)
    err_snrs_hot_2.append(err_hot2*100)
    err_snrs_hot_3.append(err_hot3*100)
    
    snr_cold0, err_cold0 = fomf.snr3d(the_imgs[i], cold_radii[0], radial_pos, cold_angles[0], 
                                bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                x_bins, y_bins, z_bins)
    snr_cold1, err_cold1 = fomf.snr3d(the_imgs[i], cold_radii[1], radial_pos, cold_angles[1], 
                                bckg_radius, phi0, phi_step, n_phi, x_size, y_size, z_size,
                                x_bins, y_bins, z_bins)
    snrs_cold_0.append(snr_cold0*100)
    snrs_cold_1.append(snr_cold1*100)
    
    err_snrs_cold_0.append(err_cold0*100)
    err_snrs_cold_1.append(err_cold1*100)
```

```python
plt.errorbar(list(range(0, n_iterations, save_every)), snrs_hot_0, yerr=err_snrs_hot_0, xerr=None,
             color=light_red, marker='s', capsize=3, label='hot sphere 0')
plt.errorbar(list(range(0, n_iterations, save_every)), snrs_hot_1, yerr=err_snrs_hot_1, xerr=None,
             color=light_blue, marker='s', capsize=3, label='hot sphere 1')
plt.errorbar(list(range(0, n_iterations, save_every)), snrs_hot_2, yerr=err_snrs_hot_2, xerr=None,
             color=blue, marker='s', capsize=3, label='hot sphere 2')
plt.errorbar(list(range(0, n_iterations, save_every)), snrs_hot_3, yerr=err_snrs_hot_3, xerr=None,
             color=green, marker='s', capsize=3, label='hot sphere 3')
plt.errorbar(list(range(0, n_iterations, save_every)), snrs_cold_0, yerr=err_snrs_cold_0, xerr=None,
             color='k', marker='s', capsize=3, label='cold sphere 0')
plt.errorbar(list(range(0, n_iterations, save_every)), snrs_cold_1, yerr=err_snrs_cold_1, xerr=None,
             color='orange', marker='s', capsize=3, label='cold sphere 1')
plt.title('SNR vs # iterations (thr=0.5 pes)')
plt.xlabel('Number of iterations')
plt.ylabel('SNR (%)')
plt.legend(loc=1);
```
