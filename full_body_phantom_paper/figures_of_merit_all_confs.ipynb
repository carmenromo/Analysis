```python
import os
import sys
import numpy as np

import matplotlib.pyplot         as plt
import mlem.mlem_reconstruct     as mr
import antea.image.fom_functions as fomf

#sys.stdout = open(os.devnull, "w") ## Prevents from printing
```

```python
path_to_mlem     = '/software/tofpet3d/lib/libmlem.so'

n_iter     = 60
save_every = 2

#sys.stdout = sys.__stdout__ ##Allows printing again

paths = ['/all_events/im_th0.25_TOF140ps_40579337coinc_iter',
         '/noTOF/im_th0.25_noTOF_40579337coinc_iter',
         '/phot_range1mm/24274077coinc/im_th0.25_TOF140ps_phot_like_phot_range1mm_24274077coinc_iter',
         '/phot_range3mm/35274951coinc/im_th0.25_TOF140ps_phot_like_phot_range3mm_35274951coinc_iter']

confs = {'Normal TOF 140':  (140, 0.25, 40579337, paths[0]),
         'Normal no TOF':   (  0, 0.25, 40579337, paths[1]),
         'Photo range 1mm': (140, 0.25, 24274077, paths[2]),
         'Photo range 3mm': (140, 0.25, 35274951, paths[3])}
```

```python
def read_images(th, tof, nc, path):
    rec = mr.MLEMReconstructor(libpath=path_to_mlem)
    reco_prefix_base = '/home/rolucar/full_body_phantom_paper/images_reco/'
    rec.prefix = reco_prefix_base + path
    return np.array([rec.read_image(i) for i in range(0, n_iter, save_every)])

from collections import namedtuple

tof_thr_nc = namedtuple('tof_thr_nc', 'tof, th, nc')
the_images = {tof_thr_nc(tof=tof, th=th, nc=nc): read_images(th, tof, nc, path) 
              for tof, th, nc, path in confs.values()}
```

```python
def imshow_slice_integrated_along_z(ax, conf, th, tof, nc, it, z_slice):
    if it < 1:
        it = 1
    image_slice = the_images[tof_thr_nc(tof=tof, th=th, nc=nc)][int(it/save_every-1),:,:,z_slice]
    im_integrated_along_z = np.sum(image_slice, axis=2).transpose()
    fig0 = ax.imshow(im_integrated_along_z, origin='lower')
    ax.set_title(conf, fontsize=14)
    ax.set_xlabel('X (mm)', fontsize=13)
    ax.set_ylabel('Y (mm)', fontsize=13)
    #fig.colorbar(fig0, ax=ax, label='Events')
```

```python
it       = 16
z_slice  = slice(25,35)
fig, axs = plt.subplots(1,4, figsize=(12, 8))
for ax, (conf, (tof, th, nc, _)) in zip(axs.flatten(), confs.items()):
    imshow_slice_integrated_along_z(ax, conf, th, tof, nc, it, z_slice)
    plt.tight_layout()
plt.show()
```

```python
z_slice = slice(25,35)    
for it in [0, 9, 29, 59]:
    fig, axs = plt.subplots(1,4, figsize=(12, 8))
    fig.suptitle(f'Iteration = {it+1}', fontsize=15)
    for ax, (conf, (tof, th, nc, _)) in zip(axs.flatten(), confs.items()):
        imshow_slice_integrated_along_z(ax, conf, th, tof, nc, it, z_slice)
    fig.tight_layout()
    fig.subplots_adjust(top=1.5)
    plt.show()
```

## Image parameters:

```python
sig_intensity  =  4
bckg_intensity =  1
radial_pos     = 50.
bckg_radius    =  4.
phi0           = np.pi/6
phi_step       = np.pi/3
n_phi          = 6

hot_radii   = [4., 6.5, 8.5, 11.]
cold_radii  = [14., 18.5]
all_radii   = hot_radii + cold_radii

hot_angles  = [np.pi/3, 2*np.pi/3, 3*np.pi/3, 4*np.pi/3]
cold_angles = [5*np.pi/3, 6*np.pi/3]

x_size = y_size = z_size = 180
x_bins = y_bins = z_bins =  60

radii  = hot_radii  + cold_radii
angles = hot_angles + cold_angles
```

```python
num_hot     = 4
num_cold    = 2
num_spheres = num_hot + num_cold

sph_thresh_tof_nc = namedtuple('sph_thresh_tof_nc', 'sph, th, tof, nc')
snr_err           = namedtuple('snr_err'          , 'snrs, errs')

def get_snrs(images, tof_thr_nc, n_sph):
    calculate_snrs_and_errs = lambda images: fomf.snr3d(images, radii[n_sph], radial_pos, angles[n_sph], 
                                                        bckg_radius, phi0, phi_step, n_phi, 
                                                        x_size, y_size, z_size,
                                                        x_bins, y_bins, z_bins)
    all_images = tuple(map(calculate_snrs_and_errs, images[tof_thr_nc]))
    snrs = [r[0] for r in all_images]
    errs = [e[1] for e in all_images]
    print(tof_thr_nc, n_sph)
    return snr_err(snrs=snrs, errs=errs)

snrs = {sph_thresh_tof_nc(sph=n_sph, th=th, tof=tof, nc=nc) :
        get_snrs(the_images, tof_thr_nc(tof=tof, th=th, nc=nc), n_sph)
        for (tof, th, nc, _) in confs.values()
        for n_sph in range(num_spheres)}
```

```python
## Configuration for plots ##
plt.rcParams["font.size"] = 14
plt.rcParams["figure.figsize"] = 10, 7.5

## colors
tableau = [( 31, 119, 180), (174, 199, 232), (255, 127,  14), (255, 187, 120),
           ( 44, 160,  44), (152, 223, 138), (214,  39,  40), (255, 152, 150),
           (148, 103, 189), (197, 176, 213), (140,  86,  75), (196, 156, 148),
           (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),
           (188, 189,  34), (219, 219, 141), ( 23, 190, 207), (158, 218, 229), 
           (218,  60,  78), (35,  180, 203)]
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau)):
    r, g, b = tableau[i]
    tableau[i] = (r / 255., g / 255., b / 255.)

light_red  = tableau[20]
light_blue = tableau[21]

## 
sph_colors=[['deepskyblue'  , light_blue,  'cornflowerblue', 'royalblue'], #sph 0 hot
            ['indianred'    , light_red,     'firebrick'   , 'orangered'], #sph 1 hot
            ['olive'        , 'yellowgreen', 'green'       , 'mediumseagreen'], #sph 2 hot
            ['orange'       , 'darkgoldenrod', 'sandybrown', 'khaki'], #sph 3 hot
            ['darkslateblue', 'navy',        'blue'        , 'royalblue'], #sph 4 cold
            ['k'            , 'silver',      'grey'        , 'lightslategrey']] #sph 5 cold

colors = {sph_thresh_tof_nc(sph=n_sph, th=th, tof=tof, nc=nc): color
          for n_sph, colors in zip(range(num_spheres), sph_colors)
          for (tof, th, nc, _), color in zip(confs.values(), colors)}
```

```python
from typing import Sequence
from typing import Dict

sel_n_iter = 60
def plot_snr_sphs(snrs:    Dict[sph_thresh_tof_nc, snr_err], 
                  spheres: Sequence[int], 
                  th:      int,
                  tof:     int,
                  nc:      int,
                  num_it:  int,
                  colors:  Dict[sph_thresh_tof_nc, 'Color'], 
                  lab:     str):
    images_wanted = int(num_it/save_every)
    for n_sphere in spheres:
        sphere = snrs  [sph_thresh_tof_nc(sph=n_sphere, th=th, tof=tof, nc=nc)]
        color  = colors[sph_thresh_tof_nc(sph=n_sphere, th=th, tof=tof, nc=nc)]
        plt.errorbar(x = list(range(0, num_it, save_every)), 
                     y =    sphere.snrs[:images_wanted], 
                     yerr = sphere.errs[:images_wanted], 
                     xerr = None,
                     color = color, 
                     marker = 's', 
                     capsize = 3,
                     label =f'sph{n_sphere}, ({all_radii[n_sphere]} rad), {lab}')
```

```python
plt.figure(figsize=(12,8))
for lab, (tof, th, nc, _) in confs.items():
    plot_snr_sphs(snrs, range(4), th, tof=tof, nc=nc, num_it=60, colors=colors, lab=lab)
plt.title(f'SNR Hot spheres', fontsize=16)
plt.xlabel('Number of iterations', fontsize=15)
plt.ylabel('SNR', fontsize=15)
plt.legend(loc=1, fontsize=13, ncol=2);
```

```python
## plt.figure(figsize=(12,8))
for lab, (tof, th, nc, _) in confs.items():
    plot_snr_sphs(snrs, range(4,6), th, tof=tof, nc=nc, num_it=60, colors=colors, lab=lab)
plt.title(f'SNR Cold spheres', fontsize=16)
plt.xlabel('Number of iterations', fontsize=15)
plt.ylabel('SNR', fontsize=15)
plt.legend(loc=4, fontsize=13, ncol=1);
```

```python
crc_err = namedtuple('crc_err', 'crcs, errs')
sig_intensity  =  4
bckg_intensity =  1

def get_crcs(images, tof_thr_nc, n_sph):
    calculate_crcs_and_errs = lambda images: fomf.crc_hot3d(images, sig_intensity, bckg_intensity, 
                                                            radii[n_sph], radial_pos, angles[n_sph], 
                                                            bckg_radius, phi0, phi_step, n_phi, 
                                                            x_size, y_size, z_size,
                                                            x_bins, y_bins, z_bins)
    all_images = tuple(map(calculate_crcs_and_errs, images[tof_thr_nc]))
    crcs = [r[0]*100 for r in all_images]
    errs = [e[1]*100 for e in all_images]
    print(tof_thr_nc, n_sph)
    return crc_err(crcs=crcs, errs=errs)

crcs = {sph_thresh_tof_nc(sph=n_sph, th=th, tof=tof, nc=nc):
        get_crcs(the_images, tof_thr_nc(tof=tof, th=th, nc=nc), n_sph)
        for (tof, th, nc, _) in confs.values()
        for n_sph in range(4)}
```

```python
def get_crcs_cold(images, tof_thr_nc, n_sph):
    calculate_crcs_and_errs = lambda images: fomf.crc_cold3d(images, radii[n_sph], radial_pos, angles[n_sph], 
                                                             bckg_radius, phi0, phi_step, n_phi, 
                                                             x_size, y_size, z_size,
                                                             x_bins, y_bins, z_bins)
    all_images = tuple(map(calculate_crcs_and_errs, images[tof_thr_nc]))
    crcs = [r[0]*100 for r in all_images]
    errs = [e[1]*100 for e in all_images]
    print(tof_thr_nc, n_sph)
    return crc_err(crcs=crcs, errs=errs)

crcs_cold = {sph_thresh_tof_nc(sph=n_sph, th=th, tof=tof, nc=nc):
             get_crcs_cold(the_images, tof_thr_nc(tof=tof, th=th, nc=nc), n_sph)
             for (tof, th, nc, _) in confs.values()
             for n_sph in range(4, 6)}
```

```python
def plot_crc_sphs(crcs:    Dict[sph_thresh_tof_nc, crc_err], 
                  spheres: Sequence[int], 
                  th:      int,
                  tof:     int,
                  nc:      int,
                  num_it:  int,
                  colors:  Dict[sph_thresh_tof_nc, 'Color'],
                  lab:     str):
    images_wanted = int(num_it/save_every)
    for n_sphere in spheres:
        sphere = crcs  [sph_thresh_tof_nc(sph=n_sphere, th=th, tof=tof, nc=nc)]
        color  = colors[sph_thresh_tof_nc(sph=n_sphere, th=th, tof=tof, nc=nc)]
            
        plt.errorbar(x = list(range(0, num_it, save_every)), 
                     y =    sphere.crcs[:images_wanted], 
                     yerr = sphere.errs[:images_wanted], 
                     xerr = None,
                     color = color, 
                     marker = 's', 
                     capsize = 3,
                     label =f'sph{n_sphere}, ({all_radii[n_sphere]} rad), {lab}')
```

```python
plt.figure(figsize=(14,10))
for lab, (tof, th, nc, _) in confs.items():
    plot_crc_sphs(crcs, range(4), th, tof=tof, nc=nc, num_it=60, colors=colors, lab=lab)
plt.title(f'CRC Hot spheres', fontsize=16)
plt.xlabel('Number of iterations', fontsize=15)
plt.ylabel('CRC', fontsize=15)
plt.legend(loc=4, fontsize=13, ncol=2)
plt.ylim(-15, 90);
```
