```python
import os
import numpy as np
import matplotlib.pyplot as plt
import antea.database.load_db      as db
```

```python
DataSiPM = db.DataSiPM('petalo', 0)
```

```python
## Define the figure
fig = plt.figure(figsize=(10,10))
plt.xlim(-220, 220)
plt.ylim(-220, 220)
plt.xlabel('X', size=18)
plt.ylabel('Y', size=18)

## Define the ring
r_int = 165
r_ext = 195
center_point = (0., 0.)
circ1 = plt.Circle(center_point, r_int, color='k', fill=False)
circ2 = plt.Circle(center_point, r_ext, color='k', fill=False)

ax = fig.gca()
ax.add_artist(circ1)
ax.add_artist(circ2)

## Plot the sensors
plt.plot(DataSiPM.X.values, DataSiPM.Y.values, '.', markersize=3)
```

```python
## Define the figure
fig = plt.figure(figsize=(10,10))
plt.xlim(-220, 220)
plt.ylim(-220, 220)
plt.xlabel('X', size=18)
plt.ylabel('Y', size=18)

## Define the ring
r_int = 165
r_ext = 195
center_point = (0., 0.)
circ1 = plt.Circle(center_point, r_int, color='k', fill=False)
circ2 = plt.Circle(center_point, r_ext, color='k', fill=False)

ax = fig.gca()
ax.add_artist(circ1)
ax.add_artist(circ2)

## Plot the sensors
plt.plot(DataSiPM.X.values, DataSiPM.Y.values, '.', markersize=3)

## Print points
point1 = np.array([92.77343749999979, -0.0, 0.0])

sipm1     = DataSiPM[DataSiPM.SensorID==2868]
sipm1_pos = np.array([sipm1.X.values, sipm1.Y.values, sipm1.Z.values]).transpose()
sipm2     = DataSiPM[DataSiPM.SensorID==2706]
sipm2_pos = np.array([sipm2.X.values, sipm2.Y.values, sipm2.Z.values]).transpose()
sipm3     = DataSiPM[DataSiPM.SensorID==2881]
sipm3_pos = np.array([sipm3.X.values, sipm3.Y.values, sipm3.Z.values]).transpose()

plt.plot(point1[0], point1[1], 'x', color='r', markersize=8)
plt.plot(sipm1_pos[0][0], sipm1_pos[0][1], 'x', color='r', markersize=8)
plt.plot(sipm2_pos[0][0], sipm2_pos[0][1], 'x', color='r', markersize=8)
plt.plot(sipm3_pos[0][0], sipm3_pos[0][1], 'x', color='r', markersize=8)

## Distances between all sensor with point1
d1 = np.linalg.norm(np.subtract(point1, sipm1_pos))
d2 = np.linalg.norm(np.subtract(point1, sipm2_pos))
d3 = np.linalg.norm(np.subtract(point1, sipm3_pos))
print(d1)
print(d2)
print(d3)
```
