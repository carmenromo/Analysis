```python
import os
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = '/data5/users/carmenromo/fastmc/data_extract_true_reco_info_full_body_exp_dist_tof/data_coincidences_phot_info_dist_tof_no_thr/'

list_coinc = []
for filename in os.listdir(filepath):
    if filename.endswith('0_2_2_4_4_2.npz'):
        my_file = filepath+filename
        d   = np.load(my_file)
        df0 = pd.DataFrame({})
        for (key, vals) in d.items():
            df0[key] = vals
        list_coinc.append(df0)
df_coinc = pd.concat(list_coinc).reset_index()
```

```python
true_x1 = df_coinc.a_true_r1*np.cos(df_coinc.a_true_phi1)
true_y1 = df_coinc.a_true_r1*np.sin(df_coinc.a_true_phi1)
true_x2 = df_coinc.a_true_r2*np.cos(df_coinc.a_true_phi2)
true_y2 = df_coinc.a_true_r2*np.sin(df_coinc.a_true_phi2)
reco_x1 = df_coinc.a_reco_r1*np.cos(df_coinc.a_reco_phi1)
reco_y1 = df_coinc.a_reco_r1*np.sin(df_coinc.a_reco_phi1)
reco_x2 = df_coinc.a_reco_r2*np.cos(df_coinc.a_reco_phi2)
reco_y2 = df_coinc.a_reco_r2*np.sin(df_coinc.a_reco_phi2)

true_x = np.concatenate((true_x1, true_x2), axis=0)
true_y = np.concatenate((true_y1, true_y2), axis=0)
reco_x = np.concatenate((reco_x1, reco_x2), axis=0)
reco_y = np.concatenate((reco_y1, reco_y2), axis=0)

true_z = np.concatenate((df_coinc.a_true_z1, df_coinc.a_true_z2), axis=0)
reco_z = np.concatenate((df_coinc.a_reco_z1, df_coinc.a_reco_z2), axis=0)

true_r = np.concatenate((df_coinc.a_true_r1, df_coinc.a_true_r2), axis=0)
reco_r = np.concatenate((df_coinc.a_reco_r1, df_coinc.a_reco_r2), axis=0)

true_phi = np.concatenate((df_coinc.a_true_phi1, df_coinc.a_true_phi2), axis=0)
reco_phi = np.concatenate((df_coinc.a_reco_phi1, df_coinc.a_reco_phi2), axis=0)

diff_x   = reco_x   - true_x
diff_y   = reco_y   - true_y
diff_r   = reco_r   - true_r
diff_phi = reco_phi - true_phi
diff_z   = reco_z   - true_z
```
